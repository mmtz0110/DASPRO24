```python
# MAtplotlib, SeaBorn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
bulan = ['January','February','March','April','May','June','July','August','September','October','November','December']
minggu1 = np.random.randint(200000, 700001, 12)
minggu2 = np.random.randint(200000, 700001, 12)
minggu3 = np.random.randint(200000, 700001, 12)
minggu4 = np.random.randint(200000, 700001, )

data = {
    'Bulan' : bulan,
    'Minggu1' : minggu1,
    'Minggu2' : minggu2,
    'Minggu3' : minggu3,
    'Minggu4' : minggu4
}
df = pd.DataFrame(data)
df['totalPengeluaran'] = df[['Minggu1','Minggu2','Minggu3','Minggu4']].sum(axis=1)
df
```

```python
plt.figure(figsize=(12,6)) #width, height -> inch
plt.bar(df['Bulan'],df['totalPengeluaran'], color='#274e13')
plt.xticks(rotation=45)
plt.tight_layout()
plt.title("Total Pengeluaran Mahasiswa")
plt.ylabel("Rupiah")
plt.grid(axis='y')
plt.show()
```

```python
#Tren pengeluaran per minggu dalam satu tahun
plt.figure(figsize=(12,6))

for minggu in ['Minggu1','Minggu2','Minggu3','Minggu4']:
    plt.plot(df['Bulan'],df[minggu],marker='o', label = minggu)

plt.title("Tren Pengeluaran Mahasiswa per Minggu")
plt.ylabel("Rupiah")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#diagram pie (proporsi pengeluaran per bulan)
plt.figure(figsize=(6,6))
plt.pie(df['totalPengeluaran'], labels=df['Bulan'],startangle=90, autopct='%1.1f%%',colors=plt.cm.Pastel2.colors)
plt.axis('equal')
plt.title("Proporsi Pengeluaran Mahasiswa Per Bulan")
plt.show()
```

```python
# Model Ai -> scikit, sklearn
from sklearn.linear_model import LinearRegression
df['bulan'] = np.arange(1,13)

x = df[['bulan']]
y = df['totalPengeluaran']

model = LinearRegression()
model.fit(x,y)

prediksi_13 = model.predict([[13]])[0]
print(f"Prediksi Pengeluaran Mahasiswa Bulan Januari: {prediksi_13:,.0f}")
plt.figure(figsize=(10,10))
plt.boxplot(df['totalPengeluaran'])
plt.show()
```
